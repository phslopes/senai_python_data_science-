# Preparando o ambiente

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega a base
df = pd.read_csv("SICRO_2023_2024_Norte.csv", encoding='latin1')

# Tema padrão Seaborn
sns.set_theme(style="whitegrid")
```

```python
df2 = df.copy()
df2 = df2['Família']
df2.drop_duplicates()
list(df2)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df[df['Família'] == 'MARTELETE'],
    x="UF",
    y="Preço.Final",
    estimator="mean",
    palette="Blues_d"
)
plt.title("Preço médio por familia")
plt.xlabel("UF")
plt.ylabel("Preço médio (Preço.Final)")
plt.show()
```

podemos comprar da familia martelete em qualquer estado, pois o valor não se altera

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df[df['Item.Líder'] == 'M0181'],
    x="UF",
    y="Preço.Final",
    estimator="mean",
    palette="Blues_d"
)
plt.title("Preço médio por item")
plt.xlabel("UF")
plt.ylabel("Preço médio (Preço.Final)")
plt.show()
```

o item M0181 tem um preço tabelado por UF, a compra do mesmo pode ser realizada de qualquer UF

# Barras

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df[df['Família'] == 'MANGOTE'],
    y="UF",
    x="Preço.Final",
    palette="Greens_d"
)
plt.title("Preço por familia")
plt.xlabel("Preço final")
plt.ylabel("UF")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df[df['TipoInsumo'] == 'Mão de obra'],
    y="UF",
    x="Preço.Final",
    palette="Greens_d"
)
plt.title("Preço por TipoInsumo")
plt.xlabel("Preço final")
plt.ylabel("UF")
plt.show()
```

# Linhas

```python
df.dropna()
```

```python
df[df['REF'] == 'DMT']
```

```python
df = df[df['REF'] != 'DMT']
df_year = df.groupby("REF")["Preço.Final"].mean().reset_index()

plt.figure(figsize=(6,4))
sns.lineplot(
    data=df_year,
    x="REF",
    y="Preço.Final",
    marker="o",
    color="crimson"
)
plt.title("Preço médio trimestral")
plt.xlabel("REF")
plt.ylabel("Preço médio por trimestre")
plt.show()
```

o meu preço medio subiu drasticamente na data do segundo semestre de 2024, podemos ter base que é apos as ferias do meio do ano.

# Histogramas

```python
df = df[df['TipoInsumo'] == 'Mão de obra']
plt.figure(figsize=(6,4))
sns.histplot(
    df['Preço.Final'],
    bins=20,
    color="steelblue"
)
plt.title("Histograma de preços")
plt.xlabel("Preço")
plt.ylabel("Frequência")
plt.show()
```

podemos baservar que o preço da mão de obra quanto mais alto o valor menor a frequancia

# Dispersão

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=df[df['TipoInsumo'] == 'Mão de obra'],
    x="Preço.Entregue",
    y="Preço.Final",
    hue="UF",
    alpha=0.7,
    palette="tab10"
)
plt.title("Preço.Entregue vs. Preço.Final")
plt.xlabel("Preço.Entregue")
plt.ylabel("Preço.Final")
plt.legend(title="Categoria")
plt.show()
```

o preço final e entregue tem um volume maior na faixa de 4000 ou seja, podemos ter uma base de qual é o valor de gasto

# BoxPlot

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    data=df,
    x="UF",
    y="Preço.Final",
    palette="Set2"
)
plt.yscale("log")
plt.title("Boxplot de preços por UF (escala log)")
plt.xlabel("UF")
plt.ylabel("Preço (USD)")
plt.show()
```

aparentemente o uf, PA,TO,RR tem um preço maior em relação aos outros

# Heatmap

```python
corr = df[["Preço.Entregue","Preço.Final"]].corr()

plt.figure(figsize=(6,5))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)
plt.title("Matriz de Correlação")
plt.tight_layout()
plt.show()
```

o preço final e o preço medio se equivalem
