# Montar um dataFrame com a classificação do tipo de variavel de cada coluna.

```python
types_df = pd.DataFrame({
    'coluna nome': amostra1.columns,
    'Tipo variavel': amostra1.dtypes
})
types_df.loc[types_df['Tipo variavel'] == 'object', 'Tipo variavel'] = 'categorica'
types_df.loc[types_df['Tipo variavel'] == 'float', 'Tipo variavel'] = 'numerica'

display(types_df)
```

```python
types_df1 = pd.DataFrame({
    'coluna nome': amostra2.columns,
    'Tipo variavel': amostra2.dtypes
})
types_df1.loc[types_df1['Tipo variavel'] == 'object', 'Tipo variavel'] = 'categorica nominal'
types_df1.loc[types_df1['Tipo variavel'] == 'float', 'Tipo variavel'] = 'numerica continuo'
types_df1.loc[types_df1['coluna nome'] == 'Relação.Liderança', 'Tipo variavel'] = 'categorica ordinal'
types_df1.loc[types_df1['Tipo variavel'] == 'int', 'Tipo variavel'] = 'numerica discreta'
display(types_df1)
```

#Nas amostras calcular a frequência absoluta e relativa (%) da coluna TipoInsumo.


```python
freq_abs = amostra1['TipoInsumo'].value_counts().sort_index()
freq_rel =  (freq_abs / len(amostra1))*100
pd.DataFrame({'Frequência absoluta': freq_abs, 'Frequência relativa (%)': freq_rel.round(2)})
```

#Calcular as medidas de tendencia central por Região, por UF e por Tipo


```python
uf = amostra1.groupby('UF')['Preço.Final'].agg(
    mean='mean',
    median='median',
    )
display(uf)

tipo_insumo = amostra1.groupby('TipoInsumo')['Preço.Final'].agg(
    mean='mean',
    median='median',
)
display(tipo_insumo)
```

```python
modauf = amostra1.groupby('UF')['Relação.Liderança'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modauf)
```

```python
modatipo = amostra1.groupby('TipoInsumo')['Família'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modatipo)
```

Com base nas medidas de tendência central calculadas, parece haver variações nos preços finais médios e medianos entre os diferentes estados e tipos de insum  na região Norte.
Por exemplo, alguns estados apresentam preços médios mais altos do que outros, e o tipo de insumo 'Equipamento' tem um preço médio muito superior em comparação com os demais, aparentemente a moda de relação de liderança é unanime em lider e a moda de familias em tipo insumo apenas tem moda em Material DMT.
